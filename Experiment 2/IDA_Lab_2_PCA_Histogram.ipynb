```python
# Importing Required Libraries
import pandas as pd
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("/content/winequality-red.csv")
print("Dataset is:\n")
print(df)
```

```python
# Here we are storing pca function in a variable
pca = PCA(n_components = 2)
# Passing Dataframe to function
pca_data = pca.fit(df).transform(df)
print("Transformed Dataset is:\n")
print(pca_data)
```

```python
# Visualizing the original dataset
sns.scatterplot(df)
# Adding title
plt.title("Scatter Plot of Original Dataset")
# Adding Labels
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
# Visualizing the reduced dataset
sns.scatterplot(pca_data)
# Adding title
plt.title("Scatter Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
```

```python
# Plotting Histogram of Dataset
sns.histplot(df)
# Adding x limit
plt.xlim(0,30)
# Adding title
plt.title("Histogram Plot of Original Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")
```

```python
# Plotting Histogram of Dataset
# Since we are directly giving dataset it, for all the comman bins of 0 and 1 columns it will represent it with an additional grey colour
sns.histplot(pca_data)
# Adding x limit
plt.xlim(-50,125)
# Adding title
plt.title("Histogram Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")
```

```python
# Plot histograms for both columns
# In the previous histogram it was showing one additional plot to remove that we will plot both columns separately
# It will also make it easier to recognize
sns.histplot(data=pca_data, x=pca_data[:,0], label="Principle Component 1")
sns.histplot(data=pca_data, x=pca_data[:,1], label="Principle Component 2")

# Adding x limit
plt.xlim(-50,125)

# Adding title
plt.title("Histogram Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")

# Show legend
plt.legend()
```

```python
# Plotting Bar Plot of Dataset
sns.barplot(df)
# Adding title
plt.title("Bar Plot of Original Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")
```

```python
# Plotting Bar Plot of Dataset
# Since there are only two columns in the reduced dataset it will directly show that
# Values are both in positive and negative
sns.barplot(x=pca_data[:,0],y=pca_data[:,1])
# Adding title
plt.title("Bar Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")
```

```python
# Trying for another dataset for student dropout dataset
df = pd.read_csv("/content/Student-Dropout-dataset.csv")
print("Dataset is:\n")
print(df)
```

```python
# First 6 rows of dataframe
print("First 6 rows of dataframe:\n")
df.head()
```

```python
# Seeing the shape of dataframe
print("Shape of Dataframe is:",df.shape)
```

```python
# Since the Target variable contains string value we need to remove it from dataset to do pca else it will give error
# But we will use this for later plotting so before removing we will store in target variable
target = df['Target']
df.drop(columns = ['Target'],axis=1,inplace = True)
pca = PCA(n_components=2)
pca_data = pca.fit(df).transform(df)
print("Transformed Dataset is:\n")
print(pca_data)
```

```python
# Now before plotting we need to do label encoding on target
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the data
target = label_encoder.fit_transform(target)

# Printing after Label encoding
print("Target value after encoding is:\n")
print(target)
```

```python
# Visualizing the original dataset
sns.scatterplot(df)
# Adding title
plt.title("Scatter Plot of Original Dataset")
# Adding Labels
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
# Visualizing the reduced dataset
# Giving color palette to seaborn
custom_palette = ['purple','yellow','green']
sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],data = pca_data,hue=target,palette=custom_palette)
# Adding title
plt.title("Scatter Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
```

```python
# Plotting Histogram of Dataset
sns.histplot(df)
# Adding x limit
plt.xlim(0,45)
# Adding title
plt.title("Histogram Plot of Original Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")
```

```python
# Plotting Histogram of Dataset
# Since we are directly giving dataset it, for all the comman bins of 0 and 1 columns it will represent it with an additional grey colour
sns.histplot(pca_data)
# Adding title
plt.title("Histogram Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")
```

```python
# Plot histograms for both columns
# In the previous histogram it was showing one additional plot to remove that we will plot both columns separately
# It will also make it easier to recognize
sns.histplot(data=pca_data, x=pca_data[:,0], label="Principle Component 1")
sns.histplot(data=pca_data, x=pca_data[:,1], label="Principle Component 2")

# Adding title
plt.title("Histogram Plot of Transformed Dataset")
# Adding Labels
plt.xlabel("Columns")
plt.ylabel("Count of Values")

# Show legend
plt.legend()
```
